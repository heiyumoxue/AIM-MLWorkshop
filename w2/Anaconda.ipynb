### If you want to change the default conda to pyenv version
**!!Please back up the ~/.bashrc file!!**
- Open the file
``
vi ~/.bashrc 
``
- Comment all the conda initialize setttings
- Add new PATH

```
export PYENV_ROOT="$HOME/.pyenv" 
export PATH="$PYENV_ROOT/bin:$PATH"  
eval "$(pyenv init --path)" 
eval "$(pyenv init -)"
```

**save and quit**``:wq``

````python
source ~/.bashrc
````
